```python
display(Math('4+3 = 7'))
```

```python
x = 4
y = 5

display(Math( str(x) + ' + ' + str(y) + ' = ' + str(x + y) ))
# display(Math(r"x = " + str(x)))
```

%g are placeholders in string format.
The following x,y replaces string values of x and y (4 + 5)
Then x+y is evaluated and put in the last %g
I hope there is a simpler way of doing this :o!

```python
display(Math('%g+%g = %g' %(x,y,x+y)))
```

```python
display(Math('4/5 = .8'))
display(Math( '\\frac{4}{5} = 0.8' ))
```

```python
x = 7
y = -2
z = 5

# Calculate the result of the expression
ans = 3 * x * (4 + y)

display(Math('3x(4+y) = %g' %ans))

# Display the equation with calculated values
# display(Math(r"3 \times %g \left( 4 + %g \right) = %g" % (x, y, result)))
display(Math( '3\\times%g(4%g) = %g' %(x,y,ans) ))
```

```python
ans = -y -(x+3)/z

display(Math('-y -\\frac{x+3}{z} =%g' %ans))
display(Math('-%g -\\frac{%g+3}{%g} =%g' %(y,x,z,ans) ))
```
